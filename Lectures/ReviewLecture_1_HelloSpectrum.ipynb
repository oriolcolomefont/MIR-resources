```python
soundsDIR = '../data/freesound/'
fs = 44100
x = ess.MonoLoader(filename = soundsDIR + 'whistle_2.wav', sampleRate = fs)()
#Normalize maximum amplitude to 1
x = x / np.max(np.abs(x))
ipd.Audio(x, rate = fs)#including the sound in the output of the cell
```

```python
#Ploting the time-domain signal
t = np.arange(x.size) / float(fs)
plt.figure(figsize = (12, 4))
plt.subplot(1, 2, 1)
plt.plot(t, x)
plt.title('Time Domain Signal')
plt.ylabel('amplitude')
plt.xlabel('time (seconds)')
plt.axis('tight')

#Your code starts here----------
#Write the code to plot the amplitude spectrum (in dBs)

#Initialize parameters
#Pick an fft size that is a power of 2 higher than the signal size

#Computing log-magnitude spectrum using fft (of scipy imported)
#before computing the log, it is preferable to set values smaller than np.finfo(float).eps to epsilon

#Plot amplitude spectrum
plt.subplot(1, 2, 2)

plt.title('Amplitude spectrum')
plt.ylabel('amplitude (dB)')
plt.xlabel('frequency (Hz)')
plt.show()
#Your code ends here----------
```
