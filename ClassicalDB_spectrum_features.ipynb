# Tonality classicalDB ConstantQ spectrum features
Audios in Tonality classicalDB dataset are proprietaries. We are going to upload to zenodo different Constant-Q spectrums for each dataset audio.

There are some parameterizations in common between the system inputs in the last five year proposals for key global detection (from 2015). That is why we are going to compute 6 different CQT spectrums with frame_size = {2048, 4096, 8192} and windows_type = {hann, blackmanharris}.

The cqt algorithm used is defined here:
https://librosa.org/doc/latest/generated/librosa.cqt.html

And implemented here:
https://librosa.org/doc/latest/generated/librosa.cqt.html

```
import librosa
import json
import os
import itertools
import numpy as np
import sys
np.set_printoptions(threshold=sys.maxsize)
```

```
os.system("mkdir /root/mir_datasets/classicalDB/spectrograms")
```

```
def save_json(dictionary, name_file):
    with open(name_file, 'w') as fp:
        json.dump(dictionary, fp, sort_keys=True, indent=4)


for k, track in classicalDB.load_tracks().items():
    print("Calculating spectrums " + track.title + "....", end='')
    y, sr = librosa.load(track.audio_path)
    spectrum = librosa.cqt(y, sr=sr, window='blackmanharris', hop_length=4096)
    spectrum = {'spectrum': [list(map(str, x)) for x in spectrum.tolist()]}
    save_json(spectrum,  os.path.splitext(track.audio_path.replace('/audio/', '/spectrums/'))[0] + '.json')
    print("OK")
```
